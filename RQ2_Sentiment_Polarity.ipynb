```python
import pandas as pd
import time
from matplotlib import pyplot as plt
from datetime import datetime
```

```python
def save(df):
    ts = int(time.time())
    df.to_excel("./files/{}.xlsx".format(ts))
    print(str(ts)+".xlsx has been saved")
```

```python
df = pd.read_csv("./data/sample.csv",delimiter=",",dtype=str)
print(df.shape)
```

# Polarity of Tweets from 2016 to 2021

```python
def year_quarter(year,quarter):
    return str(year)+"_"+str(quarter)
```

```python
temp = df.value_counts(subset=['vader_polarity'])
temp_df = pd.DataFrame(list(zip([t[0] for t in temp.index],list(temp.values))), columns = ["vader_polarity","count"])
temp_df.head()
```

```python
df["year_quarter"] = df.apply(lambda row: year_quarter(row["year"] , row["quater"]), axis = 1)
temp = df.value_counts(subset=['year_quarter','vader_polarity'])
#temp_df = pd.DataFrame(list(zip([t[0] for t in temp.index],list(temp.values))), columns = ["year_quarter","count"])
save(pd.DataFrame(temp))
```

### Polarity During Peak Weeks

```python
def year_week(year,wk):
    return str(year)+"_"+str(int(float(wk)))

df["year_week"] = df.apply(lambda row: year_week(row["year"] , row["week"]), axis = 1)
df.sample(3)
```

```python
lst_dominant_weeks = ["2021_32","2016_51","2020_46"]
df_dominant_weeks = df.query("year_week == '2016_51' | year_week == '2020_46' | year_week == '2021_32'")
df_dominant_weeks.sample(5)
```

```python
temp = df_dominant_weeks.value_counts(subset=['year_week','vader_polarity'])
save(pd.DataFrame(temp))
```
